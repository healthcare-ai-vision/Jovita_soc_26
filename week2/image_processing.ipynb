```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread("image_face.png")
img2 = cv2.imread("image_scenery.png")
```

```python
img1_rgb = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)

plt.imshow(img1_rgb)
plt.axis("off")
plt.show()
```

```python
img2_rgb = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

plt.imshow(img2_rgb)
plt.axis("off")
plt.show()
```

```python
print(img1.shape)
print(img2.shape)
```

```python
gray1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
gray2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))
plt.title("Original Image 1")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(gray1, cmap='gray')
plt.title("Grayscale Image 1")
plt.axis("off")

plt.show()
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
plt.title("Original Image 2")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(gray2, cmap='gray')
plt.title("Grayscale Image 2")
plt.axis("off")

plt.show()
```

```python
# Histogram for image 1
plt.figure(figsize=(8, 5))
plt.hist(gray1.ravel(), bins=256, range=[0, 256])
plt.title("Histogram of Grayscale Image 1")
plt.xlabel("Pixel Intensity")
plt.ylabel("Number of Pixels")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.hist(gray2.ravel(), bins=256, range=[0, 256])
plt.title("Histogram of Grayscale Image 2")
plt.xlabel("Pixel Intensity")
plt.ylabel("Number of Pixels")
plt.show()
```

```python
resized1 = cv2.resize(gray1, (300, 300))
resized2 = cv2.resize(gray2, (300, 300))
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(gray1, cmap='gray')
plt.title(f"Original ({gray1.shape[1]}x{gray1.shape[0]})")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(resized1, cmap='gray')
plt.title("Resized (300x300)")
plt.axis("off")

plt.show()
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(gray2, cmap='gray')
plt.title(f"Original ({gray2.shape[1]}x{gray2.shape[0]})")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(resized2, cmap='gray')
plt.title("Resized (300x300)")
plt.axis("off")

plt.show()
```

```python
edges1 = cv2.Canny(gray1, 100, 200)
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(gray1, cmap='gray')
plt.title("Original Grayscale Image 1")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(edges1, cmap='gray')
plt.title("Edge Detection (Canny)")
plt.axis("off")

plt.show()
```

```python
edges2 = cv2.Canny(gray2, 100, 200)
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(gray2, cmap='gray')
plt.title("Original Grayscale Image 2")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(edges2, cmap='gray')
plt.title("Edge Detection (Canny)")
plt.axis("off")

plt.show()
```

```python
noise = np.random.normal(0, 25, gray1.shape)

noisy1 = gray1.astype(np.float32) + noise
noisy1 = np.clip(noisy1, 0, 255).astype(np.uint8)
```

```python
noise = np.random.normal(0, 40, gray2.shape)

noisy2 = gray2.astype(np.float32) + noise
noisy2 = np.clip(noisy2, 0, 255).astype(np.uint8)
```

```python
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(gray1, cmap='gray')
plt.title("Original")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(noisy1, cmap='gray')
plt.title("Noisy")
plt.axis("off")

plt.show()
```

```python
denoised1 = cv2.GaussianBlur(noisy1, (5,5), 0)
```

```python
denoised2 = cv2.GaussianBlur(noisy2, (5,5), 0)
```

```python
plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
plt.imshow(gray1, cmap='gray')
plt.title("Original")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(noisy1, cmap='gray')
plt.title("Noisy")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(denoised1, cmap='gray')
plt.title("After Gaussian Filter")
plt.axis("off")

plt.show()
```

```python
plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
plt.imshow(gray2, cmap='gray')
plt.title("Original")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(noisy2, cmap='gray')
plt.title("Noisy")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(denoised2, cmap='gray')
plt.title("After Gaussian Filter")
plt.axis("off")

plt.show()
```
